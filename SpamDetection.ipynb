```python
#impoting all the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, roc_curve, confusion_matrix
from sklearn.preprocessing import LabelEncoder
```

```python
#reading the spam mail csv
df = pd.read_csv('/content/spam.csv', encoding='ISO-8859-1')
```

```python
# displaying the head of the DataFrame
df.head()
```

```python
# droping the unnecessary columns from the dataset
df = df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1)
```

```python
df.head()
```

```python
# label encoding the target variable as
# spam: 1 and ham: 0
encoder = LabelEncoder()
df['v1'] = encoder.fit_transform(df['v1'])
df.head()
```

```python
# Replacing all the missing or null values with a empty string
df.fillna('', inplace=True)
df.head()
```

```python
# seperating the dependent and independent variable
x=df.drop('v1', axis=1)
y=df['v1']
print(x.shape)
print(y.shape)
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Assuming x_train and x_test are DataFrames with a single column 'text'
train_text = x_train['v2'].tolist()
test_text = x_test['v2'].tolist()

# TF-IDF vectorization
vectorizer = TfidfVectorizer(min_df=5, stop_words='english', lowercase=True)
train_feat = vectorizer.fit_transform(train_text)
test_feat = vectorizer.transform(test_text)

print("Shape of TF-IDF matrix for training data:", train_feat.shape)
print("Shape of TF-IDF matrix for testing data:", test_feat.shape)
```

```python
print(train_feat)
# print(test_feat)
```

```python
model = LogisticRegression()
model.fit(train_feat, y_train)
y_pred = model.predict(test_feat)
y_pred_train = model.predict(train_feat)
acc_test = accuracy_score(y_test, y_pred)
acc_train = accuracy_score(y_train, y_pred_train)
report = classification_report(y_test, y_pred)
print(f'Train accuracy: {acc_train}')
print(f'Test accuracy: {acc_test}')
print(report)
```

```python
model = RandomForestClassifier()
model.fit(train_feat, y_train)
y_pred = model.predict(test_feat)
y_pred_train = model.predict(train_feat)
acc_test = accuracy_score(y_test, y_pred)
acc_train = accuracy_score(y_train, y_pred_train)
report = classification_report(y_test, y_pred)
print(f'Train accuracy: {acc_train}')
print(f'Test accuracy: {acc_test}')
print(report)
```

```python
model = SVC(C=1.5)
model.fit(train_feat, y_train)
y_pred = model.predict(test_feat)
y_pred_train = model.predict(train_feat)
acc_test = accuracy_score(y_test, y_pred)
acc_train = accuracy_score(y_train, y_pred_train)
report = classification_report(y_test, y_pred)
print(f'Train accuracy: {acc_train}')
print(f'Test accuracy: {acc_test}')
print(report)
```
